```python
def Fib(n):
    if n == 0 :
        return 0
    if n == 1 :
        return 1
    return Fib(n-1) + Fib(n-2)

Fib(5)
```

```python
n = 10
F = [None for i in range(n)] #memo
F[0] = 0
F[1] = 1
def Fib(n):
    if F[n] != None: 
        return F[n]
    else:
        f = Fib(n-1) + Fib(n-2)

    F[n] = f
    return f

print( Fib(9) )
print(F)
```

```python
#memoized version
n=5
r = [None for i in range(n+1)] #r has n+1 elements
p = [0, 1, 5, 8, 9, 10, 17, 17, 20, 24, 30]

def cut_rod(n):
    if r[n] is not None:
        return r[n]
    if n == 0 :
        q = 0
    else:
        q = float('-inf')
        for i in range(1, n+1):
            q = max (q, p[i] + cut_rod(n-i))        
        r[n] = q
    return q
print(r)    
print( cut_rod(5) )
print(r)
```

```python
#bottom-up version
n=5
r = [None for i in range(n+1)] #r has n+1 elements
p = [0, 1, 5, 8, 9, 10, 17, 17, 20, 24, 30]

def cut_rod_bottom_up(n):
    r[0] = 0
    for j in range(1, n+1):
        
        q = float('-inf')
        for i in range(1, j+1):  #i is the position of the first cut
            q = max (q, p[i] + r[j-i])   
        r[j] = q
    return r[n]
print( cut_rod_bottom_up(5) )
print(r)
```
